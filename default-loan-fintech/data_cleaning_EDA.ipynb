### DAYS_AGE, DAYS_WORK, DAYS_REGISTRATION and DAYS_ID_CHANGE

<hr>

* It's easier for us to understand these data if we converted to years

* The formula is simple, just <b>divide by -365</b>

* <b>No need to scale and binning</b>, because we'll use Tree-based model

```python
train_df['DAYS_AGE'] = (train_df['DAYS_AGE'] / -365)
test_df['DAYS_AGE'] = (test_df['DAYS_AGE'] / -365)
 
train_df['DAYS_REGISTRATION'] = (train_df['DAYS_REGISTRATION'] / -365)
test_df['DAYS_REGISTRATION'] = (test_df['DAYS_REGISTRATION'] / -365)

train_df['DAYS_ID_CHANGE'] = (train_df['DAYS_ID_CHANGE'] / -365)
test_df['DAYS_ID_CHANGE'] = (test_df['DAYS_ID_CHANGE'] / -365)

# Create an anomalous flag column
train_df['DAYS_WORK_ANOMALIES'] = train_df['DAYS_WORK'] == 365243
train_df['DAYS_WORK_ANOMALIES'] = train_df['DAYS_WORK_ANOMALIES'].astype(np.uint8)
# Replace the anomalous values with nan
train_df['DAYS_WORK'].replace({365243: np.nan}, inplace = True)
train_df['DAYS_WORK'] = (train_df['DAYS_WORK'] / -365)

test_df['DAYS_WORK_ANOMALIES'] = test_df['DAYS_WORK'] == 365243
test_df['DAYS_WORK_ANOMALIES'] = test_df['DAYS_WORK_ANOMALIES'].astype(np.uint8)
test_df['DAYS_WORK'].replace({365243: np.nan}, inplace = True)
test_df['DAYS_WORK'] = (test_df['DAYS_WORK'] / -365)
```

```python
# Since only 32 (in train) and 5 (in test) applicants who have more than 4 children
# i convert value more than 4 to 4, in other words, 4 is the maximum limit value
# this technique known as feature clipping

train_df['NUM_CHILDREN'] = train_df['NUM_CHILDREN'].apply(lambda x: 4 if x >= 4 else x)
test_df['NUM_CHILDREN'] = test_df['NUM_CHILDREN'].apply(lambda x: 4 if x >= 4 else x)
```

```python
# Make use all ORGANIZATION_TYPE in train data are exist in test data

ORGANIZATION_TYPE_TRAIN = set(train_df['ORGANIZATION_TYPE'].unique())
ORGANIZATION_TYPE_TEST = set(test_df['ORGANIZATION_TYPE'].unique())

if len(ORGANIZATION_TYPE_TRAIN.difference(ORGANIZATION_TYPE_TEST)) == 0:
    print('All ORGANIZATION_TYPE in train data are exist in test data')
```

## Little Feature Engineering

* We'll add more feature engineering on this dataset later, this is for EDA purpose

```python
# INST_DAYS: When the installment of previous credit was supposed to be paid (relative to application date of current loan)
# PAY_DAYS: When was the installments of previous credit paid actually (relative to application date of current loan)

# AMT_INST: What was the prescribed installment amount of previous credit on this installment
# AMT_PAY: What the client actually paid on previous credit on this installment
```

```python
installment_payment['PAYMENT_DIFFERENCE'] = installment_payment['AMT_INST'] - installment_payment['AMT_PAY']
installment_payment['PAYMENT_DIFFERENCE_FLAG'] = (installment_payment['PAYMENT_DIFFERENCE'] > 0).astype(int)
installment_payment['PAYMENT_PERC'] = installment_payment['AMT_PAY'] / installment_payment['AMT_INST']
installment_payment['LATE_PAYMENTS_DAYS'] = installment_payment['INST_DAYS'] - installment_payment['PAY_DAYS']
installment_payment['LATE_PAYMENTS_FLAG'] = (installment_payment['LATE_PAYMENTS_DAYS'] > 0).astype(int)
```

```python
LATE_PAYMENTS_FLAG__ = installment_payment['LATE_PAYMENTS_FLAG'].value_counts()
PAYMENT_DIFFERENCE_FLAG__ = installment_payment['PAYMENT_DIFFERENCE_FLAG'].value_counts()

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

ax[0].pie(LATE_PAYMENTS_FLAG__.values,
          labels=LATE_PAYMENTS_FLAG__.index,
          autopct='%.1f%%',
          pctdistance=1.35,
          textprops={'fontsize': 20, 'color': 'black'},
          wedgeprops={'linewidth': 0.5, 'edgecolor': 'black'},
          labeldistance=None,
          startangle=40,
          explode=[0.0, 0.1],
          colors=['#66b3ff', 'indianred'])

ax[0].axis('equal')
# ax[0].legend()
ax[0].set_title('Late Payment', fontsize=24, pad=10)

ax[1].pie(PAYMENT_DIFFERENCE_FLAG__.values,
          labels=PAYMENT_DIFFERENCE_FLAG__.index,
          autopct='%.1f%%',
          pctdistance=1.35,
          textprops={'fontsize': 20, 'color': 'black'},
          wedgeprops={'linewidth': 0.5, 'edgecolor': 'black'},
          labeldistance=None,
          startangle=0,
          explode=[0.0, 0.1],
          colors=['#66b3ff', 'indianred'])

ax[1].axis('equal')
ax[1].legend()
ax[1].set_title('Paid Less than his / her Obligation', fontsize=24, pad=10)

plt.tight_layout()
plt.show()
```
